### Nettoyage des données

```python
# Valeurs manquantes representées par "?"
# remplacer "?" par NaN
df.replace("?", np.nan, inplace= True)
```

```python
# Nombre de valeurs manquantes par colonne
df.isnull().sum()
```

D'après la description du jeu de données, les variables normalized-losses, bore, stroke, horsepower et peak-rpm
sont de variables continues. La variable num-of-doors est une variable catégorielle.
La variable price est notre variable cible donc nous allons tout simplement supprimer les valeurs manquantes de
cette colonne.

```python
# Imputation par la moyenne des colonnes normalized-losses, bore, stroke, horsepower et peak-rpm
def imput_avg(X):
    avg = X.astype('float').mean(axis =0)
    return X.replace(np.nan, avg, inplace = True)
for colonne in ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm']:
    imput_avg(df[colonne])
```

```python
# Imputation pour la variable categorielle num-of-doors
mode = df['num-of-doors'].value_counts().idxmax() 
# idxmax() unefonction pour trouver l’indice de la valeur maximale le long de l’axe de la colonne. 
df['num-of-doors'].replace(np.nan, mode, inplace = True)
```

```python
# Suppression des valeurs manquantes de la colonne cile price
df.dropna(subset=['price'], axis = 0, inplace = True)
```

```python
# Reinitialisation des indices
df.reset_index(drop = True, inplace = True)
```

```python
# Verification desimputations
# valeurs manquantes
df.isnull().sum()
```
